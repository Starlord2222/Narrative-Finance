```python
classifier = pipeline('sentiment-analysis', model="cardiffnlp/twitter-roberta-base-sentiment")
```

```python
classifier("No doubt that Bitcoin is a decentralized digital currency that uses cryptography to secure transactions, still hackers can steal bitcoins by gaining access to bitcoin owners' digital wallets.")
```

```python
sample = df1[['preprocessed']].head(100)
```
